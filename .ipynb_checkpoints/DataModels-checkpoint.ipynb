<h1>Data Models for Harvey Needs Dataset</h1>

<div>
Harvey dataset contains data for emergency needs and for shelters. Models will cluster the needs based on location and the shelters based on location. Then the sum of the Euclidean distances between cluster centroids will be used to evaluate the current locaitons of shelters.
</div>

<i>loading the data</i>

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics 

# request the needs, products, and shelters data from the API
# parse the response body as pandas DF
d1 = requests.get('https://api.harveyneeds.org/api/v1/needs').json()
n = pd.DataFrame(list(d1.items())[0][1])

#create a DF to hold locations and drop records missing info
needs = n[['latitude', 'longitude']].dropna()

d3 = requests.get('https://api.harveyneeds.org/api/v1/shelters').json()
s = pd.DataFrame(list(d3.items())[0][1])
shelters = s[['latitude', 'longitude']].dropna()
```

<h2>First the number of clusters is found using DBSCAN</h2>

```python
# first cluster needs with DBSCAN to find # of clusters needed

n2 = needs[needs['latitude'] > 29]
n3 = n2[n2['latitude'] < 31]
n4 = n3[n3['longitude'] > -99]

# assume each need is significant
# "noise" does not exist (or is infrequent) so doesnt need to be avoided
# min_samples can be low
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=.3, min_samples=3).fit(n3)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True

# Number of clusters in labels, ignoring noise if present.
dblabels = db.labels_
nClusters = len(set(dblabels)) - (1 if -1 in dblabels else 0)

plt.scatter(n3['latitude'], n3['longitude'], 4, marker=',')
plt.title("graph to be clustered")
plt.show()
```

```python
# Black removed and is used for noise instead.
dbunique_labels = set(dblabels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(dbunique_labels))]
for k, col in zip(dbunique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (dblabels == k)

    xy = n3[class_member_mask & core_samples_mask]
    plt.plot(xy['latitude'], xy['longitude'], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = n3[class_member_mask & ~core_samples_mask]
    plt.plot(xy['latitude'], xy['longitude'], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % nClusters)
plt.show()
```

<h2>
Using agglomerative hierarchical clustering to find the best cluster labeling
</h2>

```python
from sklearn.cluster import AgglomerativeClustering
nClusters = 5
ac = AgglomerativeClustering(nClusters).fit(n3) # similarity by euclidean distance, Wards method for merging clusters

# Number of clusters in labels, ignoring noise if present.
aclabels = ac.labels_
# Black removed and is used for noise instead.
acunique_labels = set(aclabels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(acunique_labels))]
for k, col in zip(acunique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (aclabels == k)

    xy = n3[class_member_mask]
    plt.plot(xy['latitude'], xy['longitude'], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=14)

    xy = n3[class_member_mask]
    plt.plot(xy['latitude'], xy['longitude'], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Number of clusters: %d' % nClusters)
plt.show()
```

<h2>Now cluster with KMeans to find centroid locations</h2>

```python
from sklearn.cluster import KMeans
km = KMeans(nClusters).fit(n3) # default to KMeans++ initialization

# Number of clusters in labels, ignoring noise if present.
kmlabels = km.labels_
# Black removed and is used for noise instead.
kmunique_labels = set(kmlabels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(kmunique_labels))]
for k, col in zip(kmunique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (kmlabels == k)

    xy = n3[class_member_mask]
    plt.plot(xy['latitude'], xy['longitude'], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=10)

    xy = n3[class_member_mask]
    plt.plot(xy['latitude'], xy['longitude'], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Number of clusters: %d' % nClusters)
plt.show()
```

<h2>Now cluster the shelters data with KMeans as well</h2>

```python
s2 = shelters[shelters['latitude'] < 35]

plt.scatter(s2['latitude'], s2['longitude'], 4, marker=',')
plt.title("graph to be clustered")
plt.show()
```

```python
km2 = KMeans(nClusters).fit(s2) # default to KMeans++ initialization

# Number of clusters in labels, ignoring noise if present.
kmlabels = km2.labels_
# Black removed and is used for noise instead.
kmunique_labels = set(kmlabels)
colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, len(kmunique_labels))]
for k, col in zip(kmunique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = (kmlabels == k)

    xy = s2[class_member_mask]
    plt.plot(xy['latitude'], xy['longitude'], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=10)

    xy = s2[class_member_mask]
    plt.plot(xy['latitude'], xy['longitude'], 'o', markerfacecolor=tuple(col),
             markeredgecolor='k', markersize=6)

plt.title('Number of clusters: %d' % nClusters)
plt.show()
```
